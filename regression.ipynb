```r
library(tidyverse)
library(tidymodels)
```

```r
nba <- read_csv("https://raw.githubusercontent.com/jakobsereda/dsci-100-project/main/stats.csv")
nba
```

```r
nba_data <- nba |>
    select(PTS, MIN, AST, REB, BLK, FT_PCT, FG_PCT, FG3_PCT, PF) |>
    na.omit()
nba_data
```

```r
nba_split <- initial_split(nba_data, prop = 0.75, strata = PTS)
nba_training <- training(nba_split)
nba_testing <- testing(nba_split)
```

```r
nba_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
    set_engine("kknn") |>
    set_mode("regression")

nba_recipe <- recipe(PTS ~ MIN + REB + AST, data = nba_training) |>
    step_scale(all_predictors()) |>
    step_center(all_predictors())

nba_recipe
```

```r
nba_vfold <- vfold_cv(nba_training, v = 6, strata = PTS)

nba_workflow <- workflow() |>
    add_recipe(nba_recipe) |>
    add_model(nba_spec) 
nba_workflow
```

```r
gridvals <- tibble(neighbors = seq(from = 1, to = 11, by = 1))

nba_results <- nba_workflow |>
    tune_grid(resamples = nba_vfold, grid = gridvals) |>
    collect_metrics() 

nba_results
```

```r
nba_min <- nba_results |>
    filter(.metric == "rmse") |>
    slice_min(std_err, n = 1)
nba_min
```

```r
k_min <- nba_min |>
    pull(neighbors)

nba_best_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = k_min) |>
    set_engine("kknn") |>
    set_mode("regression")

nba_best_fit <- workflow() |>
    add_recipe(nba_recipe) |>
    add_model(nba_best_spec) |>
    fit(data = nba_training)

nba_summary <- nba_best_fit |>
    predict(nba_testing) |>
    bind_cols(nba_testing) |>
    metrics(truth = PTS, estimate = .pred)

nba_summary
```

```r
nba_max <- nba |>
    select(PTS) |>
    na.omit() |>
    summarize(max(PTS))
nba_min <- nba |>
    select(PTS) |>
    na.omit() |>
    summarize(min(PTS))
nba_min
nba_max
```
